# MENTAL HEALTH
## PROBLEM DESCRIPTION
Mental health affects students perfomance and well-being. This task involves eight mental health survey questions to classify a student's state as either "HEALTHY", "STRESSED" or "DEPRESSED."The user will respond with YES or NO to each question.
### MVP
INPUT:Anwer the 8 questions on the survey

PROCESS:Counts the number "YES" answers.

OUTPUT:The appropriate mental health status.


# ALGORITHM

1. **Start**

2. **Initialize**:

   * Set `score ← 0`
   * Set `question_number ← 1`

3. **Display** a welcome message:

   * “Welcome to the Mental Health Survey!”

4. **Repeat** the following steps for 8 questions:

   *  Prompt the user with a question about their mental state (e.g., sleep patterns, motivation, mood, etc.)
   *  Ask the user to rate their experience:

     * 1 = Never
     * 2 = Sometimes
     * 3 = Often
     * 4 = Always
      **If** the user response is 3 or 4:

     * Increase `score` by 1

5. **After all 8 responses**:

   * a. **If** `score <= 2`, then:

     * Set `classification ← "Healthy"`
   * b. **Else if** `score <= 5`, then:

     * Set `classification ← "Stressed"`
   * c. **Else**, then:

     * Set `classification ← "Depressed"`

6. **Display** the final classification to the user:

   * “Based on your responses, you are classified as: \[classification]”

7. **End**


```python
def ask_question(question):
    response = input(question + " (yes/no): ").strip().lower()
    if response in ['yes', 'y']:
        return 1
    else:
        return 0

def classify_mental_health(score):
    if score <= 2:
        return "Healthy"
    elif score <= 5:
        return "Stressed"
    else:
        return "Depressed"

def mental_health_survey():
    print("Welcome to the Mental Health Survey")
    print("Please answer the following 8 questions honestly.\n")

    questions = [
        "Do you feel overwhelmed with school work?",
        "Do you have trouble sleeping at night?",
        "Have you lost interest in activities you used to enjoy?",
        "Do you feel hopeless or extremely sad frequently?",
        "Are you easily irritated or angry?",
        "Do you experience a loss of appetite or overeat?",
        "Do you have difficulty concentrating?",
        "Have you thought about giving up or withdrawing socially?"
    ]

    score = 0
    for q in questions:
        score += ask_question(q)

    result = classify_mental_health(score)
    print(f"\n Based on your responses, you are classified as: {result}")

mental_health_survey()
```
